```python
#Importamos las librerías necesarias:
import spotipy
import pandas as pd
from spotipy.oauth2 import SpotifyClientCredentials

#Añadimos el ID Cliente
client_id = "f075c399154f4adfa58ca82acce44ed4"

#Añadimos el valor del secret client:

client_secret = "1ca527e2ef1a4794b555ececd9f539c2"

client_credentials_manager = SpotifyClientCredentials (client_id, client_secret)

#Creamos el cliente que se va a usar para consumir la API. 

sp = spotipy.Spotify (client_credentials_manager=client_credentials_manager)
```

```python
#Buscamos información sobre latino/2022 - Bucle para buscar las 500 canciones

dicc_hiphop_2022 = {"Artista": [], "Album": [], "Cancion": [], "Año Lanzamiento": [], "Total tracks": []}
contador = 0
while contador < 500:
    for valor in range(0,451,50):
        result_hiphop_total = sp.search(q='genre: hiphop year: 2022', type='track', limit=50, offset={valor})
        for elemento in result_hiphop_total['tracks']['items']:
            dicc_hiphop_2022 ["Artista"].append(elemento['artists'][0]['name'])
            dicc_hiphop_2022["Album"].append(elemento['album']['name'])
            dicc_hiphop_2022 ["Cancion"].append(elemento['name'])
            dicc_hiphop_2022 ["Año Lanzamiento"].append(elemento['album']['release_date'])
            dicc_hiphop_2022 ["Total tracks"].append(elemento['album']['total_tracks'])
            contador += 1
print (dicc_hiphop_2022)
```

```python
df_dicc_hiphop_2022 = pd.DataFrame (dicc_hiphop_2022)
```

```python
df_dicc_hiphop_2022.to_csv ("dicc_hiphop_2022.csv")
```
