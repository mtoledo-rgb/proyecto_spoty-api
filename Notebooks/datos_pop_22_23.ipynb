```python
import os 
from dotenv import load_dotenv
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
```

```python
# Coloca tus credenciales aquí
load_dotenv()

CLIENT_ID_Z = os.getenv('CLIENT_ID_Z')
CLIENT_SECRET_Z = os.getenv('CLIENT_SECRET_Z')
```

```python
# Autenticación
auth_manager = SpotifyClientCredentials(CLIENT_ID_Z, CLIENT_SECRET_Z)
sp = spotipy.Spotify(auth_manager=auth_manager)
```

```python
dicc_pop_total_2022={'artista':[], 'cancion':[], 'tipo':[], 'nombre_album':[], 'total_tracks':[], 'año_lanzamiento':[]}

contador=0
while contador < 500:

        for valor in range(0,451,50):
         result_pop_total=sp.search(q="genre:pop year:2022",  type='track', limit=50, offset={valor})
         
         for track in result_pop_total['tracks']['items']:
            dicc_pop_total_2022['artista'].append(track['artists'][0]['name'])
            dicc_pop_total_2022['cancion'].append(track['name'])
            dicc_pop_total_2022['tipo'].append(track['type'])
            dicc_pop_total_2022['nombre_album'].append(track['album']['name'])
            dicc_pop_total_2022['total_tracks'].append(track['album']['total_tracks'])
            dicc_pop_total_2022['año_lanzamiento'].append(track['album']['release_date'])
            contador+=1
       
#Artista: {track['artists'][0]['name']} -- Cancion: {track['name']} -- Tipo: {track['type']} -- Nombre Album: {track['album']['name']} -- Total tracks: {track['album']['total_tracks']} -- Año de lanzamiento: {track['album']['release_date']} ")
dicc_pop_total_2022
```

```python
import pandas as pd
df_dicc_pop_total_2022=pd.DataFrame(dicc_pop_total_2022)
df_dicc_pop_total_2022.to_csv ('result_pop_total_2022.csv')
```

```python
dicc_pop_total_2023={'artista':[], 'cancion':[], 'tipo':[], 'nombre_album':[], 'total_tracks':[], 'año_lanzamiento':[]}

contador=0
while contador < 500:

        for valor in range(0,451,50):
         result_pop_total_2023=sp.search(q="genre:pop year:2023",  type='track', limit=50, offset={valor})
         
         for track in result_pop_total['tracks']['items']:
            dicc_pop_total_2023['artista'].append(track['artists'][0]['name'])
            dicc_pop_total_2023['cancion'].append(track['name'])
            dicc_pop_total_2023['tipo'].append(track['type'])
            dicc_pop_total_2023['nombre_album'].append(track['album']['name'])
            dicc_pop_total_2023['total_tracks'].append(track['album']['total_tracks'])
            dicc_pop_total_2023['año_lanzamiento'].append(track['album']['release_date'])
            contador+=1

dicc_pop_total_2023     
```

```python
df_dicc_pop_total_2023=pd.DataFrame(dicc_pop_total_2023)
df_dicc_pop_total_2023.to_csv ('result_pop_total_2023.csv')
```

```python
dataframe_pop=pd.concat([df_dicc_pop_total_2022, df_dicc_pop_total_2023],ignore_index=True) #para que ignore el indice que tiene actual y cree uno nuevo del 0 al 999.
```
