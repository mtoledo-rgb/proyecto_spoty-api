```python
# Se importan las librer√≠as necesarias.
import spotipy
import pandas as pd
from spotipy.oauth2 import SpotifyClientCredentials

# Uso de API's
import requests

# Importar librer√≠as para procesamiento de texto
import re

# Se declara el ID del cliente.
client_id = 'dfff11bd971645718dc18ad43117de2b'
 
# Se declara el valor del secret client.
client_secret = '5c52326b7fc84a3bb8fc8d36b939c6a3'
client_credentials_manager = SpotifyClientCredentials(client_id, client_secret)
 
# Se crea el cliente que se utilizar√° para consumir la API.
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)
```

```python
# Se busca informaci√≥n sobre latino/2022 - Bucle para buscar las 500 canciones

dicc_latino_22 = {"Artista": [], "Album": [], "Cancion": [], "A√±o Lanzamiento": [], "Total tracks": []}

contador = 0

while contador < 500:
    for valor in range(0,451,50):
        result_latino_total_22 = sp.search(q='genre: latino year: 2022', type='track', limit=50, offset={valor})

        for elemento in result_latino_total_22['tracks']['items']:
            dicc_latino_22["Artista"].append(elemento['artists'][0]['name'])
            dicc_latino_22["Album"].append(elemento['album']['name'])
            dicc_latino_22["Cancion"].append(elemento['name'])
            dicc_latino_22["A√±o Lanzamiento"].append(elemento['album']['release_date'])
            dicc_latino_22["Total tracks"].append(elemento['album']['total_tracks'])
            contador += 1

print (dicc_latino_22)
```

```python
# Transformar el diccionario de Latino/2022 en un Dataframe

df_latino_22 = pd.DataFrame(dicc_latino_22)
```

```python
# Print del dataframe

df_latino_22
```

```python
# Se busca informaci√≥n sobre latino/2023 - Bucle para buscar las 500 canciones

dicc_latino_23 = {"Artista": [], "Album": [], "Cancion": [], "A√±o Lanzamiento": [], "Total tracks": []}

contador = 0

while contador < 500:
    for valor in range(0,451,50):
        result_latino_total = sp.search(q='genre: latino year: 2023', type='track', limit=50, offset={valor})

        for elemento in result_latino_total['tracks']['items']:
            dicc_latino_23["Artista"].append(elemento['artists'][0]['name'])
            dicc_latino_23["Album"].append(elemento['album']['name'])
            dicc_latino_23["Cancion"].append(elemento['name'])
            dicc_latino_23["A√±o Lanzamiento"].append(elemento['album']['release_date'])
            dicc_latino_23["Total tracks"].append(elemento['album']['total_tracks'])
            contador += 1

print (dicc_latino_23)
```

```python
# Transformar el diccionario de Latino/2023 en un Dataframe

df_latino_23 = pd.DataFrame(dicc_latino_23)
```

```python
# Bucle for para ir iterando todos los artistas unicos que tenemos en la lista de Latino 22/23 y hacer la llamada para cada uno de ellos

dic_json = {}

for cantante in artistas_latino_total:
    url_latino = (f'https://musicbrainz.org/ws/2/artist/?query=artist:{cantante}&fmt=json')
    llamada = requests.get(url_latino)  
    print(f"La llamada a la API nos ha dado una respuesta de tipo: {llamada.status_code}")
    # dic_json.update(llamada.json())
    dic_json[cantante] = llamada.json()

dic_json.keys()
```
