Creación de la tabla para los datos de SPOTIFY

```python
# Crear primera tabla, para los datos que viene de Spotify

cnx = mysql.connector.connect(user='root', password='AlumnaAdalab',
                              host='127.0.0.1',
                              database = "musicstream")


mycursor = cnx.cursor()
try:
    mycursor.execute("""CREATE TABLE spotipy (
                     genero VARCHAR(255),
                     artista VARCHAR(255),
                     album VARCHAR(255),
                     nombre VARCHAR(255),
                     año_lanzamiento VARCHAR(255),
                     cantidad_canciones INT NULL
                     ); """)
    print(mycursor)
except mysql.connector.Error as err:
    print(err)
    print("Error Code:", err.errno)
    print("SQLSTATE", err.sqlstate)
    print("Message", err.msg)
```

Introducción de datos del genero Latino en la tabla SPOTIPY:

Introducción de datos del genero Pop en la tabla SPOTIPY: 

```python
# Camino de donde esta el archivo csv para abrirlo como Dataframe

pop_spotipy = pd.read_csv ("../Data/dataframe_pop.csv", index_col=0)
```

```python
# Insertar la columna 'genero' para poder identificar estos datos en la tabla

pop_spotipy.insert(1,"Genero","pop")
```

```python
# Eliminar la columna Unnamed y Tipo del Dataframe

pop_spotipy = pop_spotipy.drop(columns = ['Unnamed: 0'])
pop_spotipy = pop_spotipy.drop(columns = ['tipo'])
```

```python
# Cambiar nombres de las columnas del Dataframe, para tener igual que nuestra tabla en SQL

pop_spotipy = pop_spotipy.rename(columns={'total_tracks': 'cantidad_canciones', 'nombre_album': 'album', 'cancion': 'nombre' })
```

```python
# Dataframe en lista de tuples

pop_spotipy_lista = list(pop_spotipy.itertuples(index=False, name=None))
```

```python
# Para inserir estos datos: ha sido necesario iterar sobre el Dataframe y no utilizar la lista de tuples creada antes. 
# Con la iteración y la inserción de una a una fila, ha sido posible el exito. 

cnx = mysql.connector.connect(user='root', password='AlumnaAdalab',
                              host='127.0.0.1',
                              database = "musicstream")

mycursor = cnx.cursor()


for index, row in pop_spotipy.iterrows():
    sql_query = "INSERT INTO spotipy (Genero, artista, nombre, album, cantidad_canciones, año_lanzamiento) VALUES (%s, %s, %s, %s, %s, %s);"
    valores = (row['Genero'], row['artista'], row['nombre'], row['album'], row['cantidad_canciones'], row['año_lanzamiento'])
    
    print(valores) 

   
    
    try:
        mycursor.execute(sql_query,valores)
    
        cnx.commit()
        print(mycursor.rowcount, "registros insertados.")
    except mysql.connector.Error as err:
        print("Error:", err)
        print("Error Code:", err.errno)
        print("SQLSTATE:", err.sqlstate)
        print("Message:", err.msg)
   
```
