```python
import pandas as pd
import requests
import time
```

```python
# LLAMADA OK . RESULTADOS EN LILSTA DATOS_POP

df = pd.read_csv(r'c:\Users\alias\Desktop\adalab\mod2\proyecto_spoty-api\Data\artistas_rock_2022_2023.csv')# Paso 2: Iterar sobre la lista de artistas y hacer la llamada a la API

print(df.head())

datos_rock=[]
   
for artista in df['0']:
    
    # Construir la URL para cada artista
     url = f'https://musicbrainz.org/ws/2/artist/?query=artist:{artista}&fmt=json'

     # Hacer la llamada a la API
     respuesta = requests.get(url)
    
    # Imprimir el código de estado
     print(f"La llamada a la API para {artista} ha dado una respuesta de tipo: {respuesta.status_code}")
        
        # (Opcional) Procesar la respuesta JSON
     if respuesta.status_code == 200:
            datos = respuesta.json()
            # Aquí puedes hacer algo con los datos
            datos_rock.append(datos)  # Por ejemplo, imprimir los datos recibidos
     else:
            print(f"Error al buscar {artista}: {respuesta.text}")

     time.sleep(1)  
```

```python
#Iteramos por los datos obtenidos en Musicbrainz para cada uno de los artistas que le hemos proporcionado, para extraer la infomració que necesitamos de esos artistas.  
#Los apendeamos al diccionario datos_pop_brainz.
# manejamos los errores con .get

datos_rock_brainz={"artista":[], "pais_origen": [], "area_origen": [], "Fecha_ini_act": [], "Fecha_fin_act": []}

for n in datos_rock:
    try:
        # Verificar si 'artists' existe y tiene al menos un artista
        if 'artists' in n and len(n['artists']) > 0:
            # Agregar el nombre del artista
            artista = n['artists'][0].get('name', 'no data')
            datos_rock_brainz["artista"].append(artista)

            # Obtener 'area' si está disponible
            pais_origen = n['artists'][0].get('area', {}).get('name', 'no data')
            datos_rock_brainz["pais_origen"].append(pais_origen)

            # Obtener 'begin-area' si está disponible
            area_origen = n['artists'][0].get('begin-area', {}).get('name', 'no data')
            datos_rock_brainz["area_origen"].append(area_origen)

            # Obtener 'life-span' begin y ended si están disponibles
            fecha_ini = n['artists'][0].get('life-span', {}).get('begin', 'no data')
            datos_rock_brainz["Fecha_ini_act"].append(fecha_ini)

            fecha_fin = n['artists'][0].get('life-span', {}).get('ended', 'no data')
            datos_rock_brainz["Fecha_fin_act"].append(fecha_fin)
        else:
            # Si no hay artistas en la lista, usar 'no data' para todos los campos
            datos_rock_brainz["artista"].append("no data")
            datos_rock_brainz["pais_origen"].append("no data")
            datos_rock_brainz["area_origen"].append("no data")
            datos_rock_brainz["Fecha_ini_act"].append("no data")
            datos_rock_brainz["Fecha_fin_act"].append("no data")

    except KeyError:
        # Si ocurre cualquier error inesperado en las claves, manejar con 'no data'
        datos_rock_brainz["artista"].append("no data")
        datos_rock_brainz["pais_origen"].append("no data")
        datos_rock_brainz["area_origen"].append("no data")
        datos_rock_brainz["Fecha_ini_act"].append("no data")
        datos_rock_brainz["Fecha_fin_act"].append("no data")
```
