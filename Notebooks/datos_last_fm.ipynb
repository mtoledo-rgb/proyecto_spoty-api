```python
import os 
from dotenv import load_dotenv #importamos .env para introducir ahí nuestras credenciales
import pandas as pd
import requests
import time
```

```python
# Crear una función de llamada a la API de Music Brainz

def llamar_api_lastfm(url):

    # Encabezados con el token de acceso

    # Realizar una solicitud con el GET a la url proporcionada y almacenar la llamada en una "respuesta"
    time.sleep(1)
    respuesta = requests.get(url)
    
    # Imprime el código de estado de la respuesta.
    print(f"La llamada a la API nos ha dado una respuesta de tipo: {respuesta.status_code}")  
    
    if respuesta.status_code != 200:  # Comprueba si la respuesta no fue exitosa (código de estado distinto de 200).
        print(f"El motivo por el que la llamada falló es {respuesta.reason}") # Imprime la razón de la falla.
         
    else:
        return respuesta.json()  # Si la llamada fue exitosa, devuelve los datos de respuesta en formato JSON.
    
```

```python
ak_fm =	'45275fffebb8865ea08402cc6c12a3f9'
```

```python
# Función para limpiar datos de last.fm

def limpiar_datos_lastfm(path):
    df_last_fm = pd.read_csv(path,index_col=0)

    #renombramos la columna artista por Artista,para que al acceder la clave sea siempre igual.
    df_last_fm = df_last_fm.rename(columns={'artista': 'Artista' }) 
    datos_lastfm = {'artista':[],'listeners':[],'playcount':[],'biografia':[],'artistas_similares':[]}
    for artista in df_last_fm['Artista']:
        url_fm = f'http://ws.audioscrobbler.com//2.0/?method=artist.getinfo&artist={artista}&api_key={ak_fm}&format=json'
        respuesta_fm = requests.get(url_fm)
    
        try:
            datos_api = respuesta_fm.json()
            datos_lastfm['artista'].append(datos_api['artist']['name'])
            datos_lastfm['listeners'].append(datos_api['artist']['stats']['listeners'])
            datos_lastfm['playcount'].append(datos_api['artist']['stats']['playcount'])
            datos_lastfm['biografia'].append(datos_api['artist']['bio']['summary'])
            lista_similares = []
            for nombre in datos_api['artist']['similar']['artist'][0:4]:
                lista_similares.append(nombre['name'])
            datos_lastfm['artistas_similares'].append(lista_similares)
               
              
        except:
            print(f'fallo {respuesta_fm.text}')
    print(datos_lastfm)

    return datos_lastfm
```

1. POP

```python
artistas_pop_path = '../Data/dataframe_pop.csv'
datos_pop = limpiar_datos_lastfm(artistas_pop_path)
```

```python
file_path_pop = '../Data/data_pop_fm.csv'
dataframe_pop_fm = pd.DataFrame(datos_pop)
dataframe_pop_fm.to_csv(file_path_pop)
```
